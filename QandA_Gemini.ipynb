```python
# def main(questions):
#     df = pd.DataFrame(columns=['Question', 'Answer', 'Company'])
#     cnt=1
#     tot=len(questions)
#     for question in questions:
#         print(str(cnt)+"/"+str(tot))
#         file_path1 = os.path.join(directory_path,'comp_questions.txt')
#         with open(file_path1, 'r') as file:
#             done = file.readlines()
#         if question not in done:
#             print(question,"searching....🤐")
#             file_path = os.path.join(directory_path,'EVify_gem.txt')
#             with open(file_path, 'a', encoding='utf-8') as file:
#                 file.write(f"{question}\n")
#                 result = gemini(question,1)
#                 summarised_text = gemini(result,0)
#                 if summarised_text is not None:
#                     lines = summarised_text.split("\n")
#                     for line in lines:
#                         file.write("\t" + line + "\n")
#                 else:
#                     file.write("\t" +"None"+ "\n")
#                 file.write("\n\n")
#             print("Done🥹")
#             new_data = pd.DataFrame({'Question': [question], 'Answer': [summarised_text], 'Company': ['Evify']})
#             df = pd.concat([df, new_data], ignore_index=True)
#             with open(file_path1, 'a', encoding='utf-8') as file:
#                 file.write(f"{question}\n")
#         else:
#             print(question," already done😏")
#         cnt+=1


def main(questions):
    df = pd.DataFrame(columns=['Question', 'Answer', 'Company'])
    cnt = 1
    tot = len(questions)
    for question in questions:
        print(str(cnt) + "/" + str(tot))
        file_path1 = os.path.join(directory_path, 'comp_questions.txt')
        with open(file_path1, 'r') as file:
            done = file.readlines()
        if question not in done:
            print(question, "searching....🤐")
            file_path = os.path.join(directory_path, 'Evify_final.txt')
            with open(file_path, 'a', encoding='utf-8') as file:
                file.write(f"{question}\n")
                try:
                    result = gemini(question, 1)
                    summarised_text = gemini(result, 0)
                except Exception as e:  
                    print(f"Error calling gemini functions for question: {question} - {e}")
                    summarised_text = None
                if summarised_text is not None:
                    lines = summarised_text.split("\n")
                    for line in lines:
                        file.write("\t" + line + "\n")
                else:
                    file.write("\t" + "None" + "\n")
                file.write("\n\n")
            print("Done🥹")
            new_data = pd.DataFrame({'Question': [question], 'Answer': [summarised_text], 'Company': ['Evify']})
            df = pd.concat([df, new_data], ignore_index=True)
            with open(file_path1, 'a', encoding='utf-8') as file:
                file.write(f"{question}\n")
        else:
            print(question, " already done😏")
        cnt += 1
    return df
```
